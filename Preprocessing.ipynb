## making telemetry_2 in format of first one

```python
import numpy as np
telemetry_2['speed'] = telemetry_2.apply(lambda row: row['val'] if row['name'] == 'speed' else np.nan, axis=1)
telemetry_2['odometer'] = telemetry_2.apply(lambda row: row['val'] if row['name'] == 'odometer' else np.nan, axis=1)
telemetry_2['fuel_level'] = telemetry_2.apply(lambda row: row['val'] if row['name'] == 'fuel_level' else np.nan, axis=1)
```

```python
telemetry_2
```

```python
telemetry_2 = telemetry_2.drop(columns=['name', 'val'])
```

```python
telemetry_2 
```

```python
# Assuming 'telemetry_1' is the DataFrame containing the Telemetry 1 data

# Print all column names of telemetry_1
print("Columns of telemetry_1 DataFrame:")
print(telemetry_1.columns)

print("Columns of telemetry_2 DataFrame:")
print(telemetry_2.columns)
```

```python
# Perform an inner join on 'vehicle_id' and 'timestamp' to find overlaps
overlap = pd.merge(telemetry_1, telemetry_2, on=['vehicle_id', 'timestamp'], how='inner')

# Check if any overlap exists and print the result
if not overlap.empty:
    print("Overlap found between telemetry_1 and telemetry_2.")
    print(f"Number of overlapping rows: {len(overlap)}")
    print("Sample overlapping rows:")
    print(overlap.head())  # Display the first few rows of the overlap
else:
    print("No overlap found between telemetry_1 and telemetry_2.")
```

```python
# Perform an outer join on 'vehicle_id' and 'timestamp' to combine the two tables
combined_telemetry = pd.merge(telemetry_1, telemetry_2, on=['vehicle_id', 'timestamp'], how='outer')
combined_telemetry
```

```python
# Consolidate the 'speed', 'odometer', and 'fuel_level' columns
combined_telemetry['speed'] = combined_telemetry['speed_x'].combine_first(combined_telemetry['speed_y'])
combined_telemetry['odometer'] = combined_telemetry['odometer_x'].combine_first(combined_telemetry['odometer_y'])
combined_telemetry['fuel_level'] = combined_telemetry['fuel_level_x'].combine_first(combined_telemetry['fuel_level_y'])

# Drop the original '_x' and '_y' columns to clean up the DataFrame
combined_telemetry = combined_telemetry.drop(columns=[
    'speed_x', 'speed_y', 
    'odometer_x', 'odometer_y', 
    'fuel_level_x', 'fuel_level_y'
])
combined_telemetry
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by vehicle_id and calculate mean values
average_speed = combined_telemetry.groupby('vehicle_id')['speed'].mean()
average_fuel_level = combined_telemetry.groupby('vehicle_id')['fuel_level'].mean()
average_odometer = combined_telemetry.groupby('vehicle_id')['odometer'].mean()

# Normalize the values individually
normalized_speed = (average_speed - average_speed.min()) / (average_speed.max() - average_speed.min())
normalized_fuel_level = (average_fuel_level - average_fuel_level.min()) / (average_fuel_level.max() - average_fuel_level.min())
normalized_odometer = (average_odometer - average_odometer.min()) / (average_odometer.max() - average_odometer.min())

# Set the figure size
plt.figure(figsize=(12, 6))

# Define the width of the bars
bar_width = 0.25

# Set the x locations for the groups
x = np.arange(len(average_speed.index))

# Plot each normalized bar with different offsets
plt.bar(x - bar_width, normalized_speed, width=bar_width, label='Normalized Average Speed', color='blue')
plt.bar(x, normalized_fuel_level, width=bar_width, label='Normalized Average Fuel Level', color='green')
plt.bar(x + bar_width, normalized_odometer, width=bar_width, label='Normalized Average Odometer', color='orange')

# Set the title and labels
plt.title('Normalized Average Metrics by Vehicle ID')
plt.xlabel('Vehicle ID')
plt.ylabel('Normalized Values')
plt.xticks(x, average_speed.index, rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
