```python
# Normalize each individual value before calculating means
combined_telemetry['normalized_speed'] = (combined_telemetry['speed'] - combined_telemetry['speed'].min()) / (combined_telemetry['speed'].max() - combined_telemetry['speed'].min())
combined_telemetry['normalized_fuel_level'] = (combined_telemetry['fuel_level'] - combined_telemetry['fuel_level'].min()) / (combined_telemetry['fuel_level'].max() - combined_telemetry['fuel_level'].min())
combined_telemetry['normalized_odometer'] = (combined_telemetry['odometer'] - combined_telemetry['odometer'].min()) / (combined_telemetry['odometer'].max() - combined_telemetry['odometer'].min())

# Now, group by vehicle_id and calculate mean normalized values
average_normalized_speed = combined_telemetry.groupby('vehicle_id')['normalized_speed'].mean()
average_normalized_fuel_level = combined_telemetry.groupby('vehicle_id')['normalized_fuel_level'].mean()
average_normalized_odometer = combined_telemetry.groupby('vehicle_id')['normalized_odometer'].mean()

# Set the figure size
plt.figure(figsize=(12, 6))

# Define the width of the bars
bar_width = 0.25

# Set the x locations for the groups
x = np.arange(len(average_normalized_speed.index))

# Plot each normalized bar with different offsets
plt.bar(x - bar_width, average_normalized_speed, width=bar_width, label='Normalized Average Speed', color='blue')
plt.bar(x, average_normalized_fuel_level, width=bar_width, label='Normalized Average Fuel Level', color='green')
plt.bar(x + bar_width, average_normalized_odometer, width=bar_width, label='Normalized Average Odometer', color='orange')

# Set the title and labels
plt.title('Normalized Average Metrics by Vehicle ID')
plt.xlabel('Vehicle ID')
plt.ylabel('Normalized Values')
plt.xticks(x, average_normalized_speed.index, rotation=45)

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

### Why Normalize?

The variables in this dataset—**speed**, **fuel level**, and **odometer**—are on different scales:
- **Speed** (measured in miles per hour) ranges from 0 to 120.
- **Fuel level** is a percentage ranging from 0% to 100%.
- **Odometer** measures total distance traveled in miles and can range from a few hundred to tens of thousands.

These different scales make it difficult to directly compare the variables. By normalizing each value to a common scale (0 to 1), we can fairly compare them without bias. Without normalization, larger-range variables like **odometer** would dominate smaller-scale variables like **fuel level**, leading to skewed comparisons.

## Outlier Detection

```python
# Boxplot for Speed
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_telemetry['speed'])
plt.title('Boxplot of Speed')
plt.ylabel('Speed (mph)')
plt.show()

# Boxplot for Fuel Level
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_telemetry['fuel_level'])
plt.title('Boxplot of Fuel Level')
plt.ylabel('Fuel Level (%)')
plt.show()

# Boxplot for Odometer
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_telemetry['odometer'])
plt.title('Boxplot of Odometer')
plt.ylabel('Odometer (miles)')
plt.show()
```

The median mileage (represented by the line inside the box) is around 5,000 miles, indicating that 50% of the vehicles have odometer readings below this value.The higher odometer readings, indicated beyond the upper whisker, can be considered outliers. These represent vehicles that have accumulated unusually high mileage compared to the rest of the fleet.


```python
def find_outliers_iqr(data):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1  # Interquartile range

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    return outliers

# Apply to speed, fuel_level, and odometer columns
speed_outliers = find_outliers_iqr(combined_telemetry['speed'])
fuel_level_outliers = find_outliers_iqr(combined_telemetry['fuel_level'])
odometer_outliers = find_outliers_iqr(combined_telemetry['odometer'])

print("Speed Outliers:", speed_outliers)
print("Fuel Level Outliers:", fuel_level_outliers)
print("Odometer Outliers:", odometer_outliers)
```

```python
from scipy import stats

def find_outliers_zscore(data, threshold=3):
    z_scores = stats.zscore(data)
    outliers = data[(z_scores > threshold) | (z_scores < -threshold)]
    
    return outliers

# Apply to speed, fuel_level, and odometer columns
speed_outliers_z = find_outliers_zscore(combined_telemetry['speed'])
fuel_level_outliers_z = find_outliers_zscore(combined_telemetry['fuel_level'])
odometer_outliers_z = find_outliers_zscore(combined_telemetry['odometer'])

print("Speed Outliers (Z-score):", speed_outliers_z)
print("Fuel Level Outliers (Z-score):", fuel_level_outliers_z)
print("Odometer Outliers (Z-score):", odometer_outliers_z)
```

IQR: Detected significant odometer outliers, likely because it is sensitive to the range between the 25th and 75th percentiles. The IQR method focuses on the middle spread of the data, so high odometer readings fall outside this range.

Z-Score: Did not detect odometer outliers because it is based on the number of standard deviations away from the mean. Since the odometer data may have a wider distribution, the high values didn’t exceed the Z-score threshold (commonly set at ±3).

In the context of the speed data, where fluctuations make the plot harder to interpret, a rolling average (e.g., using a 7-day window) can be applied.
